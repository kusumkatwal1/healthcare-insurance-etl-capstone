##Check for Missing Values

```python
from pyspark.sql.functions import col, count, when

patient_df.select([
    count(when(col(c).isNull(), c)).alias(c)
    for c in patient_df.columns
]).show()
```

##Null Check for Required Tables

```python
required_tables = {
    "claims": claims_df,
    "subscriber": subscriber_df,
    "grpsubgrp": grpsubgrp_df
}

for name, df in required_tables.items():
    print(f"Null count for {name}")
    df.select([
        count(when(col(c).isNull(), c)).alias(c)
        for c in df.columns
    ]).show()
```

##Remove Duplicates from Required Tables

```python
clean_tables = {
    "claims": claims_clean_df,
    "disease": disease_clean_df,
    "group": group_clean_df,
    "grpsubgrp": grpsubgrp_clean_df,
    "hospital": hospital_clean_df,
    "patient_records": patient_clean_df,
    "subgroup": subgroup_clean_df,
    "subscriber": subscriber_clean_df
}

for name, df in clean_tables.items():
    total_count = df.count()
    distinct_count = df.distinct().count()
    duplicate_count = total_count - distinct_count
    print(name, "duplicate count:", duplicate_count)
```

##Remove Duplicate Records

```python
claims_clean_df = claims_clean_df.dropDuplicates()
disease_clean_df = disease_clean_df.dropDuplicates()
group_clean_df = group_clean_df.dropDuplicates()
grpsubgrp_clean_df = grpsubgrp_clean_df.dropDuplicates()
hospital_clean_df = hospital_clean_df.dropDuplicates()
patient_clean_df = patient_clean_df.dropDuplicates()
subgroup_clean_df = subgroup_clean_df.dropDuplicates()
subscriber_clean_df = subscriber_clean_df.dropDuplicates()

print("Duplicates removed successfully")
```

##Replace Nulls with NA

```python
claims_clean_df = claims_df.fillna("NA")
disease_clean_df = disease_df.fillna("NA")
group_clean_df = group_df.fillna("NA")
grpsubgrp_clean_df = grpsubgrp_df.fillna("NA")
hospital_clean_df = hospital_df.fillna("NA")
patient_clean_df = patient_df.fillna("NA")
subgroup_clean_df = subgroup_df.fillna("NA")
subscriber_clean_df = subscriber_df.fillna("NA")

print("Null values replaced with NA")
```
